## 2. Checking and Cleaning Data Quality:

**i. Null values**

**Generate a summary of count of all the null values column wise**

```python
df_CustDet.isna().sum()
```

**Drop Null values for customer_id because central tendencies for id’s is not feasible.**

```python
df_CustDet.isna().sum()
```

```python
df_CustPoliDet=df_CustPoliDet.dropna(subset=['Customer_Id'])
```

```python
df_CustPoliDet.isna().sum()
```

```python
df_CustDet.fillna(df_CustDet.mean(),inplace=True)
```

```python
df_CustDet.isna().sum()
```

```python
df_CustPoliDet.fillna(df_CustPoliDet.mean(),inplace=True)
```

```python
df_CustPoliDet.isna().sum()
```

```python
df_CustDet['Gender']=df_CustDet['Gender'].fillna(df_CustDet['Gender'].mode()[0])
df_CustDet['Vehicle Age']=df_CustDet['Vehicle Age'].fillna(df_CustDet['Vehicle Age'].mode()[0])
df_CustDet['Vehicle Damage']=df_CustDet['Vehicle Damage'].fillna(df_CustDet['Vehicle Damage'].mode()[0])
```

```python
df_CustDet.isna().sum()
```

**ii) Outliers**

```python
q1=df_CustPoliDet.quantile(0.25)
q3=df_CustPoliDet.quantile(0.75)
IQR=q3-q1
lower=q1-(1.5*IQR)
upper=q3+(1.5*IQR)
((df_CustPoliDet<lower)|(df_CustPoliDet>upper)).sum()
```
